# PCA, or Principal Component Analysis
a statistical technique used to reduce the dimensionality of a dataset while retaining as much of the variation as possible

It involves transforming a set of correlated variables into a set of uncorrelated variables, called principal components

1. standardizes the data by subtracting the mean and dividing by the standard deviation for each variable
2. Then, calculates the covariance matrix of the standardized data
3. The eigenvectors and eigenvalues of this covariance matrix are then calculated, and the eigenvectors with the highest eigenvalues represent the principal components

```python
# create a new DataFrame with the return attributes
return_df = df[['Return_5D', 'Return_1M', 'Return_3M', 'Return_6M', 'Return_1Y']]

# standardize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
return_df_scaled = scaler.fit_transform(return_df)
pca = PCA(n_components=1)

# perform PCA on the standardized data
from sklearn.decomposition import PCA
pca = PCA(n_components=1)
return_pca = pca.fit_transform(return_df_scaled)

df_2['Return Attribute'] = return_pca.flatten()
df_2
```

```python
from sklearn.decomposition import PCA

# select the profitability metrics
profit_cols = ['Gross Income Margin', 'EBITDA Margin', 'EBIT Margin', 'Net Income Margin', 'Return on Avg Total Assets', 'Return on Avg Total Equity']
df_profit = df[profit_cols]

# standardize the data to have mean=0 and variance=1
df_profit_standardized = (df_profit - df_profit.mean()) / df_profit.std()

# run PCA with one component
pca = PCA(n_components=1)
pca.fit(df_profit_standardized)

# extract the factor scores
factor_scores = pca.transform(df_profit_standardized)

# add the factor scores to df_2
df_2['Profitability Attribute'] = factor_scores
df_2
```

```python
# select the columns for valuation attribute
valuation_cols = ['Price /Sales CFY', 'Price / Earns Ratio', 'Price / Book Value']

# fill missing values with zeros
df[valuation_cols] = df[valuation_cols].fillna(0)

# standardize the data
scaled_valuation = (df[valuation_cols] - df[valuation_cols].mean()) / df[valuation_cols].std()

# perform PCA with one component
pca_valuation = PCA(n_components=1)
valuation_component = pca_valuation.fit_transform(scaled_valuation)

# add the valuation component to df_2
df_2['Valuation Attribute'] = valuation_component
df_2
```

```python
# assign weights to each dimension
w_return = 0.3
w_profitability = 0.2
w_valuation = 0.2
w_insider_buying = 0.3

# calculate the weighted score for each stock
df_2['Weighted Score'] = (
    w_return * df_2['Return Attribute'] +
    w_profitability * df_2['Profitability Attribute'] +
    w_valuation * df_2['Valuation Attribute'] +
    w_insider_buying * df_2['Insider Buying_perc']
)

# sort the stocks by weighted score in descending order
df_2 = df_2.sort_values(by=['Weighted Score'], ascending=False)

# select the top 10 stocks
top_10 = df_2.head(10)

top_10
```
